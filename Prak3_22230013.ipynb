```python
# Input dokumen
dokumen1 = "Barangnya sesuai harapan, cakep dan mulus. penjualnya juga baik dan ramah. overall, puas banget belanja disini."
dokumen2 = "Sesuai pesanan. Sayangnya pengiriman agak telat."
dokumen3 = "Puas banget belanja di sini. Barang ok dan harga juga sesuai kantong. Sip deh."
print("Dokumen: "+dokumen1)
print("Dokumen: "+dokumen2)
print("Dokumen: "+dokumen3)
# Case Folding
hasil_cf = dokumen1.lower()

print("Case Folding:\n", hasil_cf)

# Tokenizing
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
hasil_token = tokenizer.tokenize(hasil_cf)
print("\nTokenizing:\n", hasil_token)

# Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
hasil_stemm = stemmer.stem(hasil_cf)
print("\nHasil Stemming:\n", hasil_stemm)

# Tokenisasi ulang hasil stemming
token_stemm = tokenizer.tokenize(hasil_stemm)

# Ambil daftar stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
stop_factory = StopWordRemoverFactory()
stopwords = stop_factory.get_stop_words()

# Hapus stopwords dari hasil stemming
hasil_final = [word for word in token_stemm if word not in stopwords]
print("\nHasil Setelah Stopword Removal:\n", hasil_final)
```

```python
# Import library yang dibutuhkan
from nltk.tokenize import RegexpTokenizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Inisialisasi tokenizer, stemmer, dan stopword remover
tokenizer = RegexpTokenizer(r'\w+')
stem_factory = StemmerFactory()
stemmer = stem_factory.create_stemmer()
stop_factory = StopWordRemoverFactory()
stopwords = stop_factory.get_stop_words()

# List dokumen
dokumen_list = [
    "Barangnya sesuai harapan, cakep dan mulus. penjualnya juga baik dan ramah. overall, puas banget belanja disini.",
    "Sesuai pesanan. Sayangnya pengiriman agak telat.",
    "Puas banget belanja di sini. Barang ok dan harga juga sesuai kantong. Sip deh."
]

# Proses tiap dokumen
for i, dokumen in enumerate(dokumen_list, start=1):
    print(f"\n====== Dokumen {i} ======")
    print("Original:\n", dokumen)

    # Case Folding
    hasil_cf = dokumen.lower()
    print("\nCase Folding:\n", hasil_cf)

    # Tokenizing
    hasil_token = tokenizer.tokenize(hasil_cf)
    print("\nTokenizing:\n", hasil_token)

    # Stemming
    hasil_stemm = stemmer.stem(hasil_cf)
    print("\nStemming:\n", hasil_stemm)

    # Tokenisasi ulang hasil stemming
    token_stemm = tokenizer.tokenize(hasil_stemm)

    # Stopword Removal
    hasil_final = [word for word in token_stemm if word not in stopwords]
    print("\nSetelah Stopword Removal:\n", hasil_final)
```
